### **Geração de função que cria gráfico a partir da cidade desejada**

```python
def gera_grafico(cidade):
  try:
    dados = populacao_pb.loc[populacao_pb['Localidade'] == cidade]
    dados = dados.T.iloc[2:]
    dados.columns = ['População']
    dados['População'] = dados['População'].astype(int)

    y = np.arange(len(dados.index))
    plt.figure(figsize=(10, 6))
    plt.barh(y, np.array(dados['População']), color='green', label='População Total')

    plt.xlabel('População')
    plt.ylabel('Faixa Etária')
    plt.title(f'Distribuição da População Total por Faixa Etária da cidade de {cidade}')
    plt.legend()
    plt.grid(axis='x')
    plt.yticks(y, dados.index)

    valor_maximo = max(dados['População'])
    plt.xlim(0, valor_maximo*1.1)
    for i, value in enumerate(dados['População']):
            plt.text(value + 20, i, f'{value}', va='center', ha='left', color='black')


  except:
    print('Dados não encontrados')
```

### **Visualização dos gráficos por cidade**

```python
gera_grafico('Monteiro')
```

```python
gera_grafico('João Pessoa')
```
